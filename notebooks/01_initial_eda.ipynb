```python
import pandas as pd
```

```python
df = pd.read_csv("../data/Telco-Customer-Churn.csv")
```

```python
partial_df = df[df['Contract'] == 'Two year']
print(df.info())
print(df.dtypes)
catalog = df['TotalCharges']  #partial_df['PaymentMethod']

counts = catalog.value_counts()
print(counts)
print(f"Sum of counts: {counts.sum()}")

counts = catalog.value_counts(normalize=True)
print(counts)
```

```python
import matplotlib.pyplot as plt


# 1. Prepare the data (assuming 'df' is loaded)

labels = counts.index.tolist()


# 2. Create the plot
plt.figure(figsize=(6, 8)) # Set the size of the plot
plt.pie(
    counts,
    labels=labels,
    # This format displays the percentage with 1 decimal place followed by the % sign
    autopct='%0.1f%%',
    startangle=90,
    # The 'width' parameter determines the size of the hole
    wedgeprops={'width': 0.4, 'edgecolor': 'white'}
)

plt.title('Partner Status Percentage')
plt.show()
```
